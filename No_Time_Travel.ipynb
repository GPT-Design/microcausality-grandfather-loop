```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, Function, Integral, oo, I, exp
from sympy import simplify
#from pyfeyn.diagrams import FeynDiagram
#from pyfeyn.points import Vertex
#from pyfeyn.lines import Fermion, Photon
```

B. Import SymPy and Define Symbols

```python
# Define symbolic variables (momentum, energies, times)
p1, p2, k = symbols('p1 p2 k', real=True)  # particle momenta
t, t_prime = symbols('t t_prime', real=True)  # times
E, E_prime = symbols('E E_prime', real=True)  # energies

# Interaction strength constants
g = symbols('g', real=True)

# Imaginary unit (I), time ordering, and exponentials are key in path integrals
```

C. Define the standard (non-paradox) interaction amplitude symbolically
The normal collision amplitude (without paradox):

```python
# Example amplitude for normal interaction (symbolic integral form)
amplitude_normal = Integral(g**2 * exp(-I*E*(t - t_prime)), (t_prime, -oo, t))
```

D. Now define paradoxical loop amplitude clearly (FTL / backward time):
A paradoxical backward-time interaction changes the amplitude dramatically:

```python
# Paradox amplitude example (FTL/backward-time loop):
# Time integral reversed, causing loop-back
amplitude_paradox = Integral(g**2 * exp(-I*E*(t_prime - t)), (t_prime, t, -oo))
```

```python
# Evaluate/simplify normal amplitude integral
amplitude_normal_eval = simplify(amplitude_normal.doit())
print("Normal Amplitude (evaluated):", amplitude_normal_eval)

# Evaluate/simplify paradox amplitude integral
amplitude_paradox_eval = simplify(amplitude_paradox.doit())
print("Paradox Amplitude (evaluated):", amplitude_paradox_eval)
```
